```python
#Importing scaler to normalize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df)
```

```python
df.head()
```

```python
X = df.drop(["median_house_value"],axis = 1)
Y = df[["median_house_value"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn .linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.svm import SVR

reg = SVR()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test, y_pred)
```

```python
from sklearn.neural_network import MLPRegressor
y_train = y_train.values.ravel()
reg = MLPRegressor(hidden_layer_sizes=(5, 4,4), random_state=1, max_iter=200)
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test, y_pred)
```

```python
best_acc = 0
best_comb = [0]*3
for i in range(3,11):
    for j in range(3,11):
        for k in range(3,11):
            reg = MLPRegressor(hidden_layer_sizes=(i, j,k), random_state=1, max_iter=200)
            reg.fit(x_train,y_train)
            y_pred = reg.predict(x_test)
            score = r2_score(y_test, y_pred)
            if score>best_acc:
                best_comb = [i,j,k]
                best_acc = score
            print(f"\r{(i*100+j*10+k)/(1110)*100}%                                                             ",end = '')
print()
print(best_acc,best_comb)
```

```python
from sklearn.model_selection import cross_val_score

reg = LinearRegression()
scores = cross_val_score(reg, x_train, y_train, cv=10)
```
